```python
import cv2
import numpy as np
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe' 
# Load the handwriting image
image = cv2.imread('adhd.PNG')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to create a binary image
_, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)

# Find contours of the text regions
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)




# Apply OCR to recognize text
recognized_text = pytesseract.image_to_string(gray)

# Split the recognized text into lines
lines = recognized_text.split('\n')

# Initialize lists to store line spacing and word spacing values
line_spacing_values = []
word_spacing_values = []

# Iterate through lines to calcprint(lines)
for i in range(len(lines)):
    current_line = lines[i]
    #print("Current Line : ", current_line)
    previous_line = lines[i - 1]

    # Calculate line spacing by measuring the vertical distance between lines
    line_spacing = abs(len(current_line) - len(previous_line))
    line_spacing_values.append(line_spacing)

    # Split lines into words
    current_words = current_line.split()
    previous_words = previous_line.split()
    
    # Calculate word spacing by measuring the horizontal distance between words
    if current_words and previous_words:
        word_spacing = abs(len(current_words[0]) - len(previous_words[-1]))
        word_spacing_values.append(word_spacing)
        #print("Word Spacing : ", word_spacing_values)
# Calculate average line spacing and word spacing
average_line_spacing = np.round(sum(line_spacing_values) / len(line_spacing_values), 2)
average_word_spacing = np.round(sum(word_spacing_values) / len(word_spacing_values), 2)








# Initialize variables to store margin distance and line spacing
right_margin_distance = 0 #float('inf')  # Initialize with a large value
line_spacing = 0

# Iterate through each contour (assuming each contour corresponds to a line of text)
for contour in contours:
    # Get the bounding box of the contour
    x, y, w, h = cv2.boundingRect(contour)

    # Calculate margin distance (distance from the left edge of the page to the text)
    margin_distance = w

    # Calculate line spacing
    if line_spacing == 0:
        line_spacing = h  # Initialize line spacing with the height of the first line

    # Update margin_distance if it's smaller than the previous minimum
    if margin_distance > right_margin_distance:
        right_margin_distance = margin_distance

# Print the calculated margin distance and line spacing
print("Right Margin Distance:", right_margin_distance)
print("Line Spacing:", line_spacing)


# Print the results
print(f"Average Line Spacing: {average_line_spacing} pixels")
print(f"Average Word Spacing: {average_word_spacing} pixels")
```
